```python
import requests
import json

url_book = 'https://api.douban.com/v2/book/'
url_search = 'https://api.douban.com/v2/book/search'
```

```python
keyword = 'python'
start_number = 0
search_payload = {'q': keyword,
                 'start': start_number,
                 'count': 100}
books = []
while 1:
    res = requests.get(url_search, params=search_payload)
    if res.json()['count'] != 0:
        for book in res.json()['books']:
            books.append(book)
        search_payload['start'] += 100
        print(search_payload)
    else:
        break
    #if start_number <= 40:
    #    break
```

```python
# save text
with open('douban.txt', 'w') as f:
    f.write('id^title^pubdate^publisher^pages^price^rating\n')
    for book in books:
        f.write(book['id'] + '^' + \
                book['title'] + ' ' + book['subtitle'] + '^' + \
                book['pubdate'] + '^' + \
                book['publisher'] + '^' + \
                book['pages'] + '^' + \
                book['price'] + '^' + \
                book['rating']['average'] + \
                '\n')
```
