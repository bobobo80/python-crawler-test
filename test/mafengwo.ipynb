```python
import requests
```

```python
from fake_useragent import UserAgent

ua = UserAgent()
```

```python
# 设置fake-agent
headers = {
    'user-agent': ua.chrome
}

# 创建Session，获取cookie
s = requests.Session()

# 三亚首页
r = s.get('http://www.mafengwo.cn/travel-scenic-spot/mafengwo/10030.html', headers=headers)
r.text
```

```python
# 获取游记的ajax api
url = 'http://www.mafengwo.cn/gonglve/ajax.php?act=get_travellist'

# 设置参数
place_id = 10030
page_number = 1

payload = {
    'mddid': place_id,
    'pageid': 'mdd_index',
    'sort': 1,
    'cost': 0,
    'days': 0,
    'month': 0,
    'tagid': 0,
    'page': page_number
}

# post
r = s.post(url, headers=headers, data=payload)

# list中为html的文本，解析其中的item，获取各个游记的链接
r.text
```

```python
r.content
```

```python
import json

result = r.json()

log_list = result['list']
```

```python
from bs4 import BeautifulSoup

log_item = BeautifulSoup(log_list, 'lxml')
log_item_list = log_item.select('.tn-item')

print(len(log_item_list))

for item in log_item_list:
    log_bs_obj = item.select('.tn-wrapper')[0].select('dt > a')[0]
    print(log_bs_obj.text, log_bs_obj['href'])
    
```
