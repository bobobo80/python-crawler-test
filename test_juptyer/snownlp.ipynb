```python
logs = logs_col.find({'keywords': {'$exists': True}})

print('当前计数：', logs.count())

all_keywords = {}

for log in logs:
    for key in log['keywords']:
        if key in all_keywords:
            all_keywords[key] += 1
        else:
            all_keywords[key] = 1
```
